# 1D Solution for Point Source

General solution for the diffusion is:

$$ 
\frac{\partial T}{\partial t} = \alpha \frac{\partial^{2}T}{\partial x^{2}}
$$

One solution for such system is:

$$
T = \frac{1}{2\sqrt{\pi \alpha t}}exp\Big[ -\frac{x^{2}}{4 \alpha t}\Big]
$$

which can be viewed in the animation below, showing the diffusion from a point source temperature $T_0$ at time $t=0$, resulting a Direct-delta function at $x=0$. For the ease of computation the time close to 0 will be implemented instead. 

```python
from JSAnimation.IPython_display import display_animation
import matplotlib.animation as animation
import numpy as np
import matplotlib.pyplot as plt  
%matplotlib inline

# Allocating value of alpha
alpha = 0.05

# Initiate the plot
fig = plt.figure()
ax = fig.add_subplot(111, autoscale_on=False, xlim=(-3, 3), ylim=(0, 4))
ax.grid()
line, = ax.plot([], [], color='r', ls='-', lw=3)
time_template = 'time = %.1fs'
time_text = ax.text(0.05, 0.9, '', transform=ax.transAxes)

# Define function to calculate temperature
def T(x, t):
    T = (1/(2*np.sqrt(np.pi*alpha*t)))*np.exp(-pow(x,2)/(4*alpha*t))
    return T

# Generate arrays for spatial, temperature and time
time = np.linspace(0.1, 10, 200) 
x = np.linspace(-3, 3, 200) 
for t in time:
    if t == time[0]:
        temp = T(x,t)
    else:
        temp = np.vstack([temp, T(x,t)])

# Animation!!
def animate(i):
    #print i
    line.set_data(x, temp[i])
    time_text.set_text(time_template%(time[i]))
    return line, time_text
    
def init():
    line.set_data([], [])
    time_text.set_text('')
    return line, time_text

ani = animation.FuncAnimation(fig, animate, frames=len(time), interval=100, blit=True, init_func=init)

display_animation(ani, default_mode='once')
```
